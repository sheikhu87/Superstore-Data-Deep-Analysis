```python
# Discrete data (unique values are countable)
col_list = ["Region","Quantity","Segment","Ship Mode"]
for i in col_list:
    plt.figure(figsize=(5,5))
    sns.countplot(data=df,x=i)
    plt.title(f"{i}Countplot")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#continues data (unique values are not countable)
col_list = ["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(10,10))
    plt.hist(data=df,x=i)
    plt.title(f"{i} Histogram")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    #plt.show()
```

```python
#continues data (unique values are not countable)
col_list = ["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(7,5))
    sns.displot(data=df,x=i)
    plt.title(f"{i} DisplayPlot")
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#continues data (unique values are not countable)
col_list = ["Profit","Sales","Discount"]
for i in col_list:
    plt.figure(figsize=(7,7))
    sns.boxplot(data=df,x=i)
    plt.title(f"{i} BoxPlot")
    plt.xlabel(i)
    #plt.ylabel("Frequency")
    plt.show()
```

```python
#Greater than equal to median % of customers
perc_of_higher_median=round((len(df[df['Discount']>=df['Discount'].median()])/len(df))*100)
avg_higher_med_disc_profit=df['Profit'][df['Discount']>=df['Discount'].median()].mean()
print('% of customer with Greater than equal to median discount ',perc_of_higher_median)
print('Average profit of Greater than equal to median discount ',round(avg_higher_med_disc_profit))
```

```python
#Less than median % of customers
perc_of_lower_median=round((len(df[df['Discount']<df['Discount'].median()])/len(df))*100)
avg_lower_med_disc_profit=df['Profit'][df['Discount']<df['Discount'].median()].mean()
print('% of customer with Less than median discount ',perc_of_lower_median)
print('Average profit of Lower than median discount ',round(avg_lower_med_disc_profit))
```

```python
#Continious  Data(Unique values are uncountable)
#Kernel Density estimation plot
col_list = ['Profit', 'Sales','Discount']
for i in col_list:
    plt.figure(figsize=(10,10))
    sns.kdeplot(data=df,x=i)
    plt.title(f'{i} Kde Plot')
    plt.xlabel(i)
    plt.show()
```

```python
df[i].value_counts()
```

```python
#Discrete Data Unique values are countable.
col_list = ['Region','Quantity','Segment','Ship Mode']
for i in col_list:
    plt.figure(figsize=(7,7))
    plt.pie(df[i].value_counts(), labels= df[i].value_counts().index, autopct='%.1f%%')
    plt.title(f'{i} Pie Chart')
    plt.show()
```

```python
#Continious  Data(Unique values are uncountable)
#Violin Density estimation plot
col_list = ['Profit', 'Sales','Discount']
for i in col_list:
    plt.figure(figsize=(5,5))
    sns.violinplot(data=df,x=i)
    plt.title(f'{i} Violin Plot')
    plt.xlabel(i)
    plt.show()
```
